## Define numpy array in .1 steps

```python
import numpy as np

arr = np.arange(0.0, 10, 0.1)
print(arr)
```

## Plotting quadratic function

```python
import matplotlib.pyplot as plt

x = arr
y = np.power(arr, 2)

plt.plot(x, y)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.show()
```

## Adding Noise

```python
ynoise =  np.random.normal(size=x.size)
ydata = y + ynoise
plt.plot(x, ydata, 'b-', label='data')
plt.show()
```

## Scipy's curve fitting 

```python
from scipy.optimize import curve_fit

def curve(x, a):
    return a * pow(x,2)
popt, pcov = curve_fit(curve, x, ydata)
```

```python
plt.plot(x, ydata, linewidth=2)
plt.plot(x, curve(x, popt[0]), 'r-')
```
